```python
def fix_date(x):
    """
    Fixes dates which are in 20xx
    """
    year = x.split('/')[2]
    if int(year) <= 19:
        return x[:-2] + '20' + year
    else:
        return x[:-2] + '19' + year
```

```python
def parsing_columns(df, columns):
    """
    To get name from the selected column. Columns must be in list. 1 value can be selected just use '[]'
    """
    for column in columns:
        listed = []
        for i in df[column]:
            try:
                a = literal_eval(i.strip('[]'))
                listed.append(a)
            except AttributeError:
                listed.append('NaN')
            except SyntaxError:
                listed.append('NaN')

        coba = []
        for i in listed:
            if type(i) == str:
                coba.append('NaN')
            elif type(i) == tuple:
                tes = []
                for x in i:
                    if x['name'] != '':
                        tes.append(x['name'])  
                tes2 = ', '.join(tes)
                coba.append(tes2)
            else:
                if type(i) == dict:
                    coba.append(i['name'])

        df[column] = coba
    return df
```

```python
def name_from_job(df, columns, name):
    """
    To get name according the specific role/jobs. Columns must be in list. 1 value can be selected just use '[]'
    """
    for column in columns:
        listed = []
        for i in df[column]:
            try:
                a = literal_eval(i.strip('[]'))
                listed.append(a)
            except AttributeError:
                listed.append('NaN')
            except SyntaxError:
                listed.append('NaN')

        coba = []
        for i in listed:
            if type(i) == str:
                coba.append('NaN')
            elif type(i) == tuple:
                tes = []
                for x in i:
                    for b, c in x.items():
                        if c == name:      
                            if x['name'] != '':
                                tes.append(x['name'])  
                tes2 = ', '.join(tes)
                coba.append(tes2)
            else:
                if type(i) == dict:
                    coba.append(i['name'])

        df[column + '_' + name] = coba
    return df
```

```python
#Find list of dummies of unique job
def dummies_jobs(movies):
  d = []
  for i in movies['crew']:
      if pd.isna(i):
          d.append({})
      else:
          d.append(literal_eval(i))

  jobs = []
  for i in d:
      if i == {}:
          jobs.append({})
      else:
          for k in i:
              for x, y in k.items():
                  if x == 'job':
                      jobs.append(y)

  listed_jobs = []
  for i in jobs:
      if i not in listed_jobs:
          if i != {}:
              listed_jobs.append(i)

  #Making dataframe for unique jobs 
  df_jobs = pd.DataFrame()
  for u in listed_jobs:
      v = []
      for i in d:
          if type(i) == dict:
              v.append(0)
          else:
              ada = 0
              for k in i:
                  for x, y in k.items():
                      if x == 'job':
                          if y == u:
                              ada += 1
              v.append(ada)
      df_jobs[u] = v

  return df_jobs 
```

```python
#Find list of dummies for unique genres name
def dummies_genres(movies):
  d = []
  for i in movies['genres']:
      if pd.isna(i):
          d.append({})
      else:
          d.append(literal_eval(i))

  genres = []
  for i in d:
      if i == {}:
          genres.append({})
      else:
          for k in i:
              for x, y in k.items():
                  if x == 'name':
                      genres.append(y)

  listed_genres = []
  for i in genres:
      if i not in listed_genres:
          if i != {}:
              listed_genres.append(i)

  #Making dataframe for unique genres 
  df_genres = pd.DataFrame()
  for u in listed_genres:
      v = []
      for i in d:
          if type(i) == dict:
              v.append(0)
          else:
              ada = 0
              for k in i:
                  for x, y in k.items():
                      if x == 'name':
                          if y == u:
                              ada += 1
              v.append(ada)
      df_genres[u] = v

  return df_genres
```

```python
#Find list of dummies for unique production_companies name
def dummies_companies(movies):
  d = []
  for i in movies['production_companies']:
      if pd.isna(i):
          d.append({})
      else:
          d.append(literal_eval(i))

  companies = []
  for i in d:
      if i == {}:
          companies.append({})
      else:
          for k in i:
              for x, y in k.items():
                  if x == 'name':
                      companies.append(y)

  listed_companies = []
  for i in companies:
      if i not in listed_companies:
          if i != {}:
              listed_companies.append(i)

  #Making dataframe for unique production_companies 
  df_companies = pd.DataFrame()
  for u in listed_companies:
      v = []
      for i in d:
          if type(i) == dict:
              v.append(0)
          else:
              ada = 0
              for k in i:
                  for x, y in k.items():
                      if x == 'name':
                          if y == u:
                              ada += 1
              v.append(ada)
      df_companies[u] = v

  return df_companies
```

```python
#Find list of dummies for unique production_countries name
def dummies_countries(movies):
  d = []
  for i in movies['production_countries']:
      if pd.isna(i):
          d.append({})
      else:
          d.append(literal_eval(i))

  countries = []
  for i in d:
      if i == {}:
          countries.append({})
      else:
          for k in i:
              for x, y in k.items():
                  if x == 'name':
                      countries.append(y)

  listed_countries = []
  for i in countries:
      if i not in listed_countries:
          if i != {}:
              listed_countries.append(i)

  #Making dataframe for unique production_countries
  df_countries = pd.DataFrame()
  for u in listed_countries:
      v = []
      for i in d:
          if type(i) == dict:
              v.append(0)
          else:
              ada = 0
              for k in i:
                  for x, y in k.items():
                      if x == 'name':
                          if y == u:
                              ada += 1
              v.append(ada)
      df_countries[u] = v

  return df_countries
```

```python
def RMSE(y_obs, y_pred):
    n = len(y_obs)
    rmse = np.sqrt( 1/n*np.sum((y_pred-y_obs)**2) )
    return rmse

def RMSLE(y_obs, y_pred):
    n = len(y_obs)
    rmsle = np.sqrt( 1/n*np.sum((np.log(y_pred)-np.log(y_obs))**2) )
    return rmsle
```

```python
class StackingAveragedModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, base_models, meta_model, n_folds=5):
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    # We again fit the data on clones of the original models
    def fit(self, X, y):
        self.base_models_ = [list() for x in self.base_models]
        self.meta_model_ = clone(self.meta_model)
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        
        # Train cloned base models then create out-of-fold predictions
        # that are needed to train the cloned meta-model
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models)))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                instance = clone(model)
                self.base_models_[i].append(instance)
                instance.fit(X[train_index], y[train_index])
                y_pred = instance.predict(X[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
                
        # Now train the cloned  meta-model using the out-of-fold predictions as new feature
        self.meta_model_.fit(out_of_fold_predictions, y)
        return self
   
    #Do the predictions of all base models on the test data and use the averaged predictions as 
    #meta-features for the final prediction which is done by the meta-model
    def predict(self, X):
        meta_features = np.column_stack([
            np.column_stack([model.predict(X) for model in base_models]).mean(axis=1)
            for base_models in self.base_models_ ])
        return self.meta_model_.predict(meta_features)
```
